# Importing dataset

```python
import pandas as pd
df=pd.read_csv("E:/datasetml.csv")
df
```

# Reading test data

```python
y=pd.read_csv("E:/test.csv")
y
```

# Using naive bayes for predicting test data

```python
dfi=df.iloc[:,:-1]
dfo=df['class']
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(dfi,dfo)

predicted= model.predict(y) 
print("Predicted Value:", predicted)
```
